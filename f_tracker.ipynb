```python
# Raw Package
import numpy as np
import pandas as pd

import yahoo_fin.stock_info as si

import plotly.express as px
import plotly.graph_objects as go

import statistics

pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

```python
ticker = 'MU'

trail_pe = statistics.mean([float(i) for i in si.get_stats_valuation(ticker).loc[2].values[1:]])
resize = trail_pe
print(resize)

# get earnings history
earnings = pd.DataFrame.from_dict(si.get_earnings_history(ticker))

# get price data
data = si.get_data(ticker, start_date=min(earnings.startdatetime), interval = "1mo")

fig = go.Figure()
fig.add_trace(go.Scatter(x=data.index, y=data.close,
                    mode='lines',
                    name='price'))

fig.add_trace(go.Scatter(x=earnings.startdatetime, y=resize*earnings.epsestimate,
                    mode='lines',
                    name='eps estimate'))

# fig.add_trace(go.Scatter(x=earnings.startdatetime, y=resize*earnings.epsactual,
#                     mode='lines',
#                     name='eps actual'))

fig.update_layout(
    title=f"Price vs EPS ({ticker})"
)
```
